### we will create dict to hold alternate titles

```python
# this adds to clean movies function

def clean_movie(movie):
    movie = dict(movie)         # non destructive copy
    alt_titles = {}             # dict to hold titles
    
    # loop thru list of alternate keys
    
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        # lets check if key exists in movie and if so
        # Remove key-value pair and add to alternative titles
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)                       # pop() is used for removing key from movie dict
            
    # when we pop the key we will add it to movie object
    
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles
            
    return movie
```

### In this section we want to replace key words like 'Directed by' to 'Director'

```python
# code used for this inside clean_movies
# first define change_column
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    alt_titles = {}
    # combine alternate titles into one list
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune-Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles

    # merge column names
    def change_column_name(old_name, new_name):
        if old_name in movie:
            movie[new_name] = movie.pop(old_name)
    change_column_name('Adaptation by', 'Writer(s)')
    change_column_name('Country of origin', 'Country')
    change_column_name('Directed by', 'Director')
    change_column_name('Distributed by', 'Distributor')
    change_column_name('Edited by', 'Editor(s)')
    change_column_name('Length', 'Running time')
    change_column_name('Original release', 'Release date')
    change_column_name('Music by', 'Composer(s)')
    change_column_name('Produced by', 'Producer(s)')
    change_column_name('Producer', 'Producer(s)')
    change_column_name('Productioncompanies ', 'Production company(s)')
    change_column_name('Productioncompany ', 'Production company(s)')
    change_column_name('Released', 'Release Date')
    change_column_name('Release Date', 'Release date')
    change_column_name('Screen story by', 'Writer(s)')
    change_column_name('Screenplay by', 'Writer(s)')
    change_column_name('Story by', 'Writer(s)')
    change_column_name('Theme music composer', 'Composer(s)')
    change_column_name('Written by', 'Writer(s)')

    return movie
```

## In this section, we will dive deeper into RegEx

```python
# Box office amount, --$365.6, million and --US$1,531,489
# lets do the match
# 1. Dollar sign will be maintained orescape it == \$
# 2. Add non zero digits \d specify digits, + modifier to capture 1 or more digits == "\$\d+"
# 3. Add optional decimal point, regular so its escaped \. then add ? modifier == "\$\d+\.?"
# 4. We specify digits after decimal \d but use * modifier because there might be no digit after dec == "\$\d+\.?\d*"
# 5. Add white space \s but add a modifier * since there may be more than one == "\$\d+\.?\d*\s*"
# 6. Add million or Billion, we only match by first letter, [mb]illion == "\$\d+\.?\d*\s*[mb]illion"

#lets create a variable to hold our regex -- r makes sure its a string
form_one = r'\$\d+\.?\d*\s*[mb]illion'
```

```python
# we will count how many box office values match our form
# we will use str.contains and add a flag--re.ignorecase the .sum() count numbers that return TRUE  ----== 3896 match this
box_office.str.contains(form_one, flags=re.IGNORECASE).sum()
```

```python
# lets match $1,531,489
# 1. escape $ ==\$
# 2. match one or more digits == \$\d{1,3} -- curly matches 1-3 digits
# 3. escape a comma , and add 3 digits (,\d{3})+ and also ?: to signify non-capturing group

# therefore we will have
form_two = r'\$\d{1,3}(?:,\d{3})+'
```
